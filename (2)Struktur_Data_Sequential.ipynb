A list is a data type in Python that can store various types of data, including numbers, strings, or even other objects (high flexibility).

List characteristics:

<ol>
  <li>It is denoted by square brackets [ ] during initialization, and its values are separated by commas</li>
  <li>Elements in a list can be modified or updated (<i>changeable</i>)</li>
  <li>Lists allow duplicate values</li>
  <li>Elements in a list are ordered according to the assigned label/index (not alphabetically or numerically)</li>
</ol>

Dictionary characteristics :
<ol>

##üé†<b>9. Looping Dictionary</b>

Looping can also be performed on a dictionary using the  `.items()`, Execute the following code:

```
for key, value in dict.items():
    ...
```

example :

```
# Variable initialization
skill_needed = dict(
    data_analyst = ['SQL', 'Business Knowledge', 'Descriptive Analysis', 'Diagnostics Analysis'],
    data_engineer = ['Database Knowledge', 'Data Modelling', 'Programming Lang. (Python, Scala, etc)'],
    data_scientist = ['Machine Learning', 'Business Knowledge', 'Predictive Analysis']
)

# Looping process
for key, value in skill_needed.items():
    print(f'Pada kunci = {key} mempunyai nilai = {value}')
```

# üìö **More Iterables : Tuple dan Set**

Besides lists and dictionaries, there are two other data types called tuple and set.

<h2>Tuple</h2>
A tuple is similar to a list but is defined using parentheses ( ) and the elements in a tuple cannot be updated (immutable). Example of a tuple:

```
contoh_tuple = ('TRX20230701', 'TRX20230702', 'TRX20230703', 'TRX20230704')
```

<h2>Set</h2>
A set, as the name suggests, is a collection data type that follows set properties. Elements in a set are unordered (indexing does not apply) and do not allow duplicate elements.

```
contoh_set = {'TRX20230701', 'TRX20230702', 'TRX20230703', 'TRX20230704'}
```
