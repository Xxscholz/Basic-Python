## ‚è∞ **4. Statement IF - ELIF - ELSE**

* The <b>if</b> function is a condition used to determine what action should be taken based on a specific situation.
* Statement <code>if</code> placed at the beginning and will always be executed when the program is-*running*.
* the use of <b>if</b> - <b>elif</b> - <b>else</b> can be adjusted according to the desired conditions.
* The general form of the if function is as follows: :

```
    if(first condition):
        statement to be executed if the condition is met
    elif(second condition):
        statement to be executed if the condition is met
     ...

    else:
        default statement to be executed if none of the above conditions are met
```

Note : <i> When a condition is met, the subsequent conditions will not be checked. If it is not met, the next condition will be checked, and so on. If none of the conditions are met, the else condition (default condition) will be executed (this applies sequentially)</i>

## ‚è∞ **5. Nested IF**

Nested if is the use of a condition inside another condition (a condition within a condition). The general form of a nested if is as follows:

    if(first condition):
        if (condition 1 after the first condition is met):
            statement to be executed if the condition is met
        elif(condition 2 after the first condition is met):
            statement to be executed if the condition is met
        ...
        else:
            statement to be executed if the condition is met

    elif(second condition):
        statement to be executed if the condition is met </br>
     ...

    else:
        default statement to be executed if none of the above conditions are met</br>

## ü§ñ **8. Default Argumen**

* A default argument is a value assigned to a function parameter when no value is provided by the function caller for that parameter.
* When the function is called, if the caller does not provide a value for a parameter that has a default argument, the default value will be used.

```
def tampilkan_pelanggan(cust_name = 'Anonymous'):
    print(f'Nama pelanggan = {cust_name}')
```

When the function is called but the cust_name argument is not initialized, the default value will be used.

```
tampilkan_pelanggan()

output :
    Nama pelanggan = Anonymous
```

However, if the cust_name argument is given a value, the default argument will be replaced.

```
tampilkan_pelanggan(cust_name = 'Aksara')

output :
   Nama pelanggan = Aksara
```

Therefore, it can be concluded that if an argument does not have a default value, it must be provided (mandatory). If a default argument is set, its status changes from mandatory to optional.

### üë®‚Äçüíª Task 9

Create a function named rataan_data with an argument named data (this argument is a numeric list data type). Example:

```
list_data = [1, 2, 6, 8, 4, 3]

mean = rataan_data(data = list_data)

print(f'Rataan dari data {list_data} adalah {mean}')

output :     
    Rataan dari data [1, 2, 6, 8, 4, 3] adalah 4
```

````python
# Write your code below here & exec!
def rataan_data(data):
  total = 0
  for i in data:
    total += i
  return total/len(data)
````

````python
list_data = [1, 2, 6, 8, 4, 3]

mean = rataan_data(data = list_data)

print(f'Rataan dari data {list_data} adalah {mean}')
````
