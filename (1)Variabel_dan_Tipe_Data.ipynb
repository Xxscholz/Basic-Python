Conclusion:

In Task 2-c, we learn that when using the .format() method in Python, the number of {} placeholders must match the number of variables provided. If multiple placeholders are defined but only one variable is supplied, Python will raise an error because it does not have enough values to substitute into the placeholders. This highlights the importance of ensuring consistency between placeholders and the variables passed to .format() when formatting strings.

# ‚û°Ô∏è 3. Assignment

Assignment is the act of giving a value or data to a variable.
* The variable being assigned must be placed on the left side of the equals sign (=)
* If the same variable is assigned more than once, the last assigned value will be stored
* Python supports assigning a single value to multiple variables
* Python also supports assigning multiple values to multiple variables simultaneously
* The expression on the right-hand side is evaluated first before being assigned to the variable(s) on the left-hand side

  \begin{equation}
  CSAT =  \frac{Total\ Respon\ Positif}{Total\ Respon}*100
  \end{equation}
